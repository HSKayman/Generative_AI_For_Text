```python
from pandas import read_csv
import numpy as np
from keras import Model
from keras.layers import Layer
import keras.backend as K
from keras.layers import Input, Dense, SimpleRNN
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.metrics import mean_squared_error
```

```python
# Prepare data
def generate_fibonacci_sequence(num_elements, scale_data=True):
#def generate_fibonacci_sequence(num_elements, scale_data=False):
    # Get the Fibonacci sequence
    fibonacci_sequence = np.zeros(num_elements)
    fib_n1 = 0.0
    fib_n = 1.0 
    for i in range(num_elements):
            fibonacci_sequence[i] = fib_n1 + fib_n
            #fib_n1 = fib_n
            fib_n1 = 2
            fib_n = fibonacci_sequence[i] 
    data_scaler = []
    if scale_data:
        data_scaler = MinMaxScaler(feature_range=(0, 1))
        fibonacci_sequence = np.reshape(fibonacci_sequence, (num_elements, 1))
        fibonacci_sequence = data_scaler.fit_transform(fibonacci_sequence).flatten()        
    return fibonacci_sequence, data_scaler
def generate_fibonacci_dataset(total_fib_numbers, time_steps, train_percent, scale_data=True):
#def generate_fibonacci_dataset(total_fib_numbers, time_steps, train_percent, scale_data=False):
    data, data_scaler = generate_fibonacci_sequence(total_fib_numbers, scale_data)    
    Y_indices = np.arange(time_steps, len(data), 1)
    Y = data[Y_indices]
    num_rows_x = len(Y)
    X = data[0:num_rows_x]
    for i in range(time_steps-1):
        temp = data[i+1:num_rows_x+i+1]
        X = np.column_stack((X, temp))
    # random permutation with fixed seed   
    random_generator = np.random.RandomState(seed=13)
    indices = random_generator.permutation(num_rows_x)
    split_index = int(train_percent*num_rows_x)
    train_indices = indices[0:split_index]
    test_indices = indices[split_index:]
    trainX = X[train_indices]
    trainY = Y[train_indices]
    testX = X[test_indices]
    testY = Y[test_indices]
    trainX = np.reshape(trainX, (len(trainX), time_steps, 1))    
    testX = np.reshape(testX, (len(testX), time_steps, 1))
    return trainX, trainY, testX, testY, data_scaler
```

```python
# Set up parameters
time_steps = 20
#time_steps = 3
hidden_units = 2
#epochs = 30
#epochs = 50
epochs = 500
```

```python
# Create a traditional RNN network
def create_RNN_model(hidden_units, dense_units, input_shape, activation):
    rnn_model = Sequential()
    rnn_model.add(SimpleRNN(hidden_units, input_shape=input_shape, activation=activation[0]))
    rnn_model.add(Dense(units=dense_units, activation=activation[1]))
    rnn_model.compile(loss='mse', optimizer='adam')
    return rnn_model

rnn_model = create_RNN_model(hidden_units=hidden_units, dense_units=1, input_shape=(time_steps,1), 
                   activation=['tanh', 'tanh'])

# Generate the dataset for the network
#trainX, trainY, testX, testY, data_scaler  = generate_fibonacci_dataset(1200, time_steps, 0.7)
trainX, trainY, testX, testY, data_scaler  = generate_fibonacci_dataset(2000, time_steps, 0.95)
#trainX, trainY, testX, testY, data_scaler  = generate_fibonacci_dataset(10, time_steps, 0.8)
#trainX, trainY, testX, testY, data_scaler  = generate_fibonacci_dataset(50, time_steps, 0.8)
#trainX, trainY, testX, testY, data_scaler  = generate_fibonacci_dataset(20, time_steps, 0.8)
#trainX, trainY, testX, testY, data_scaler  = generate_fibonacci_dataset(20, time_steps, 0.95)
# Train the network

#trainX, trainY
```

```python
# Add attention layer to the deep learning network
class AttentionLayer(Layer):
    def __init__(self,**kwargs):
        super(AttentionLayer,self).__init__(**kwargs)

    def build(self,input_shape):
        self.attention_weights=self.add_weight(name='attention_weight', shape=(input_shape[-1],1), 
                               initializer='random_normal', trainable=True)
        self.attention_bias=self.add_weight(name='attention_bias', shape=(input_shape[1],1), 
                               initializer='zeros', trainable=True)        
        super(AttentionLayer, self).build(input_shape)

    def call(self,x):
        # Alignment scores. Pass them through tanh function
        alignment_scores = K.tanh(K.dot(x,self.attention_weights)+self.attention_bias)
        # Remove dimension of size 1
        alignment_scores = K.squeeze(alignment_scores, axis=-1)   
        # Compute the weights
        alpha = K.softmax(alignment_scores)
        # Reshape to tensorFlow format
        alpha = K.expand_dims(alpha, axis=-1)
        # Compute the context vector
        context_vector = x * alpha
        context_vector = K.sum(context_vector, axis=1)
        return context_vector
```

```python
def create_RNN_with_attention(hidden_units, dense_units, input_shape, activation):
    input_layer=Input(shape=input_shape)
    rnn_layer = SimpleRNN(hidden_units, return_sequences=True, activation=activation)(input_layer)
    attention_layer = AttentionLayer()(rnn_layer)
    output_layer=Dense(dense_units, trainable=True, activation=activation)(attention_layer)
    rnn_attention_model=Model(input_layer,output_layer)
    rnn_attention_model.compile(loss='mse', optimizer='adam')  
    return rnn_attention_model 
```

```python
# Create the model with attention, train and evaluate
rnn_attention_model = create_RNN_with_attention(hidden_units=hidden_units, dense_units=1, 
                                  input_shape=(time_steps,1), activation='tanh')
rnn_attention_model.summary()    
#rnn_attention_model.fit(trainX, trainY, epochs=epochs, batch_size=1, verbose=2)
rnn_attention_model.fit(trainX, trainY, epochs=epochs, verbose=2)
# Evaluate model
train_error_with_attention = rnn_attention_model.evaluate(trainX, trainY)
test_error_with_attention = rnn_attention_model.evaluate(testX, testY)

# Print error
print("Train set MSE with attention = ", train_error_with_attention)
print("Test set MSE with attention = ", test_error_with_attention)
```
